```python
%%time
import dask
from dask.graph_manipulation import bind
import time
# generate random floating point values
from random import seed
from random import random

seed(1)
DATA = []

@dask.delayed
def inc(x):
    time.sleep(random()*10)
    return x + 1

@dask.delayed
def add_data(x):
    time.sleep(random()*10)
    DATA.append(x)

@dask.delayed
def sum_data(x):
    time.sleep(random()*10)
    return sum(DATA) + x

a = inc(1)
b = add_data(a)
c = inc(3)
d = add_data(c)
e = inc(5)
f = bind(sum_data, [b, d])(e)
f.compute()
```
